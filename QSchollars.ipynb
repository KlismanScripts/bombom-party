# **1.- ¿Cuál es el vector de estado de un sistema compuesto por 3 qubits , estando los qubits individuales(e independientes) en los estados  $|+\rangle $ , $|+\rangle $,$| -\rangle $,? Luego hacer las mediciones parciales para el qubit central, repetir lo mismo con qiskit.**



El vector de estado de un sistema compuesto por múltiples qubits se puede representar utilizando el producto tensorial de los vectores de estado individuales de los qubits. Dado que tienes tres qubits en estados |+⟩, |+⟩, |−⟩, podemos representar sus vectores de estado como:

|+⟩ ⊗ |+⟩ ⊗ |−⟩

El producto tensorial se realiza componente por componente, por lo que puedes calcularlo como:

|+⟩ ⊗ |+⟩ ⊗ |−⟩ = [1 1] ⊗ [1 1] ⊗ [0 -1]


```python
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.visualization import plot_histogram
from qiskit.providers.aer import AerSimulator

# Crea un circuto cuántico con 3 qubits
qc = QuantumCircuit(3)

# Aplicar la compuerta X (NOT gate) al segundo qubit
qc.x(1)

# Realizar mediciones en la base computacional
qc.measure_all()

# Simular el circuito
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit).result()

# Mostrar los resultados
counts = result.get_counts()
plot_histogram(counts)


```

# **3.- ¿Cómo se escribiría en notación de Dirac una compuerta CU3 desarrollalo e implementalo en qiskit para $\theta = \pi$, $\phi = \frac{\pi}{2}$ y $\lambda = \frac{7\pi}{2}$**

```python

import numpy as np
from qiskit import QuantumCircuit, transpile, assemble, Aer, execute
from qiskit.extensions import UnitaryGate

# Crear un circuito cuántico de 2 qubits
circuit = QuantumCircuit(2)

# Valores de los parámetros
theta = np.pi
phi = np.pi / 2
lambda_val = 7 * np.pi / 2

# Definir la compuerta U3 con los parámetros dados
u3_gate = UnitaryGate([[np.cos(theta / 2), -np.exp(1j * lambda_val) * np.sin(theta / 2)],
                       [np.exp(1j * phi) * np.sin(theta / 2), np.exp(1j * (phi + lambda_val)) * np.cos(theta / 2)]])

# Aplicar la compuerta de control con la compuerta U3 definida
circuit.append(u3_gate.control(), [0, 1])

# Visualizar el circuito
print(circuit)

# Simular el circuito
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit, simulator)
result = job.result()
statevector = result.get_statevector()

print("Estado final del estado cuántico:")
print(statevector)
```

# **# 4.- A partir de la relación existente entre la operación (módulo 2) y el operador XOR ($\oplus$). Mostrar:**

# \begin{align}
\sum_{i=1}^n x_iz_i \pmod 2=x_1z_1\oplus x_2z_2\oplus\cdots\oplus x_nz_n
\end{align}

```python


from qiskit import QuantumCircuit, Aer, execute

# Crear un circuito cuántico de n qubits, donde n es el número de términos en la suma
n = 4  # Puedes cambiar n según tu necesidad
circuit = QuantumCircuit(n, n)

# Definir los valores de x_i y z_i como qubits (|0> o |1>)
x_qubits = [0, 1, 0, 1]  # Puedes cambiar los valores según tu necesidad
z_qubits = [1, 0, 1, 0]  # Puedes cambiar los valores según tu necesidad

# Aplicar las compuertas X y Z según los valores de x_i y z_i
for i in range(n):
    if x_qubits[i] == 1:
        circuit.x(i)
    if z_qubits[i] == 1:
        circuit.z(i)

# Aplicar la compuerta CNOT (control-X) entre cada qubit de x y su correspondiente qubit de z
for i in range(n):
    circuit.cx(i, n - 1 - i)

# Medir todos los qubits
circuit.measure(range(n), range(n))

# Simular el circuito
simulator = Aer.get_backend('qasm_simulator')
job = execute(circuit, simulator, shots=1)
result = job.result()
counts = result.get_counts()

# Obtener el resultado en bits clásicos
measurement = list(counts.keys())[0]

# Calcular el resultado de la operación (módulo 2) en bits clásicos
sum_mod2_result = int(measurement, 2) % 2

# Calcular el resultado del operador XOR en bits clásicos
xor_result = int(x_qubits[0])
for i in range(1, n):
    xor_result ^= int(x_qubits[i])

# Comparar los resultados
print("Resultado de la operación (módulo 2):", sum_mod2_result)
print("Resultado del operador XOR:", xor_result)
```

# **5.- Implementa el código para la función oráculo balanceado con $n=3$, la cual retorna el circuito cuántico qc. Considere que $f(x)$ está dado por:**

# \begin{align}
f(001)&=0\\
f(000)&=0\\
f(111)&=0\\
f(110)&=0\\
f(011)&=1\\
f(010)&=1\\
f(100)&=1\\
f(101)&=1\\
\end{align}

```python
from qiskit import QuantumCircuit, Aer, execute

def balanced_oracle_3(qc, input_qubits, output_qubit):
    # Implementar la función oráculo balanceado
    qc.cx(input_qubits[0], output_qubit)
    qc.cx(input_qubits[1], output_qubit)
    qc.cx(input_qubits[2], output_qubit)

# Crear un circuito cuántico con 4 qubits (3 de entrada y 1 de salida)
input_qubits = [0, 1, 2]  # Índices de los qubits de entrada
output_qubit = 3  # Índice del qubit de salida

qc = QuantumCircuit(4, 1)

# Aplicar compuertas X según los valores de entrada
qc.x(input_qubits[0])
qc.x(input_qubits[1])

# Aplicar el oráculo balanceado
balanced_oracle_3(qc, input_qubits, output_qubit)

# Medir el qubit de salida
qc.measure(output_qubit, 0)

# Simular el circuito
simulator = Aer.get_backend('qasm_simulator')
job = execute(qc, simulator, shots=1)
result = job.result()
counts = result.get_counts()

# Mostrar los resultados
print("Resultados:", counts)
```

Sin embargo, ¿es realmente necesario aplicar $CX$ a todos los $n$ primeros qubits, o podríamos prescindir de algunos y obtener igualmente un oráculo balanceado?**

En una construcción de un oráculo balanceado utilizando la compuerta de control-X (CX), no es necesario aplicar CX a todos los $n$ primeros qubits para lograr un oráculo balanceado. En realidad, para obtener un oráculo balanceado, es suficiente con aplicar la compuerta CX a una cantidad par de qubits de control. Puedes seleccionar cualquier conjunto de $n$ qubits de control con una cantidad par de qubits activados.

Este enfoque aprovecha el hecho de que la compuerta CX cambia el estado del qubit objetivo solo si todos los qubits de control están en estado $|1\rangle$. Si tienes una cantidad par de qubits de control activados, entonces habrá una cantidad igual de 0 y 1 en la entrada, lo que hará que la función $f(x)$ sea balanceada.

Por lo tanto, se puede seleccionar un subconjunto par de qubits de control para aplicar la compuerta CX y lograr un oráculo balanceado. Esto proporciona flexibilidad en la implementación y no es necesario aplicar CX a todos los $n$ qubits de control para lograr el objetivo deseado.


# **7.- Implemente el algoritmo de Grover para determinar la máxima cantidad de aristas en un grafo de 5 nodos y 6 aristas:**

# \begin{align}
(0,3)\\
(0,4)\\
(1,3)\\
(1,4)\\
(2,3)\\
(2,4)\\
\end{align}

Consideremos el siguiente grafo con 5 nodos y 6 aristas:

\[
\begin{align*}
(0,3) \\
(0,4) \\
(1,3) \\
(1,4) \\
(2,3) \\
(2,4)
\end{align*}
\]

Queremos determinar la máxima cantidad de aristas en este grafo utilizando el algoritmo de Grover.

El algoritmo de Grover busca la combinación de aristas que cumple con cierta condición de búsqueda, en este caso, la máxima cantidad de aristas. Para hacer esto, diseñamos un oráculo que evalúa si una combinación de aristas es válida. Luego, aplicamos el algoritmo de Grover para encontrar la combinación que hace que el oráculo devuelva 1.


```python
import numpy as np
from qiskit import QuantumCircuit, transpile, assemble, Aer, execute
from qiskit.visualization import plot_histogram
from math import pi

# Definir la función oráculo
def oracle(circuit, qubits):
    circuit.cz(qubits[0], qubits[1])
    circuit.cz(qubits[0], qubits[2])
    circuit.cz(qubits[1], qubits[3])
    circuit.cz(qubits[1], qubits[4])

# Definir el algoritmo de Grover
def grover_algorithm():
    qr = QuantumRegister(5, name="q")
    oracle_qc = QuantumCircuit(qr)
    oracle(oracle_qc, qr)
    grover_circuit = QuantumCircuit(qr)

    # Inicialización uniforme de los qubits
    grover_circuit.h(qr)

    # Número de iteraciones
    num_iteraciones = int(np.floor((pi/4) * np.sqrt(2**5)))

    # Aplicar el algoritmo de Grover
    for _ in range(num_iteraciones):
        grover_circuit.append(oracle_qc.to_instruction(), qr)
        grover_circuit.h(qr)
        grover_circuit.x(qr)
        grover_circuit.h(qr)
        grover_circuit.append(oracle_qc.to_instruction(), qr)
        grover_circuit.h(qr)
        grover_circuit.x(qr)
        grover_circuit.h(qr)

    grover_circuit.measure_all()
    return grover_circuit

# Ejecutar el algoritmo de Grover
simulator = Aer.get_backend("aer_simulator")
grover_circuit = grover_algorithm()
job = execute(grover_circuit, simulator, shots=1000)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)
```

# **8.- En el algoritmo de Grover, determine el número de veces que debe ejercutarse el par $U_\omega U_f$ para que el estado final de nuestro circuito coincida con el Ket de estados deseados sabiendo que $x\in\{0,1\}^5$ y hay 4 estados que sastifacen las características de búsqueda.**


En el algoritmo de Grover, el número de veces que se debe aplicar el operador $U_\omega U_f$ se calcula aproximadamente como:

\[
\text{Número de Iteraciones} \approx \sqrt{\frac{N}{M}}
\]

Donde:
\begin{align*}
N & : \text{Número total de elementos (estados posibles)} \\
M & : \text{Número de elementos que satisfacen la condición de búsqueda}
\end{align*}

En este caso, $N = 2^5 = 32$ y $M = 4$ (4 estados que satisfacen la búsqueda). Por lo tanto:

\[
\text{Número de Iteraciones} \approx \sqrt{\frac{32}{4}} = 4
\]





```python
import math

# Número total de elementos y número de elementos que satisfacen la condición de búsqueda
N = 2**5
M = 4

# Calcular el número de iteraciones aproximado
num_iteraciones = int(math.sqrt(N/M))

# Imprimir el resultado
print(f"Número de Iteraciones aproximado: {num_iteraciones}")

```
